<h1>Introduction</h1>

<h4>What is League of Legends?</h4>
<p>
    League of Legends (LoL) is a 3D, third person MOBA (Multiplayer Online Battle Arena) game developed by Riot Games. Released in October 2009, League recently has started it's first franchise year within North America, and currently has over 81 million players daily thoughout the world. Within the franchise year, there are some notable names such as Golden Guardians (Golden State Warriors), 100Thieves (Chicago Bulls), and Echo Fox (Rich Fox). 
</p>
<h4>Gameplay</h4>
<p>
The premise of the game is simple: each game consists of two teams of 5 players, with one team labeled as red, and the other labeled as blue. Both team's primary objective is to destroy the opposing team's Nexus, which are located at the opposite sides of the map. Within the game, there are other side objectives to accumulate gold or experience to have an edge over the enemy team. Some of these side objectives are Dragon, Baron Nashor, Turrets, and Inhibitors. There are more than the four noted, but the common objective that they all share is that they are a point of pressure to draw the enemy towards your team to start a teamfight.
</p>
<h4>Bans</h4>
<p>
Banning champions is done before the game begins with each team given 5 bans to remove certain champions from the game. These champions may be excluded from the game because they percieved as OP (overpowered), unfair to play against, or to give an advantage for the match.
Each team is given 5 bans for over 140 champions to pick, and after the ban phase, each player is required to pick one of the remaining ~130 champions that form some team compositons. 
    Full details can be found here (<a href="https://na.leagueoflegends.com/en/">https://na.leagueoflegends.com/en/</a>).
</p>
<h4>Oracle's Elixer</h4>
<p>
A platform called Oracle's Elixir created by Tim "Magic" Sevenhuysen has collected multiple seasons worth of data for each professional match that has occurred within the past couple years. Along with this data release (<a href="http://oracleselixir.com/match-data/">http://oracleselixir.com/match-data/</a>), he revealed that numerous coaches and analysts within the scene uses this data to gain a better idea at what the strengths and weaknesses are for each team and player. Given the interesting statistics they were able to find, the Oracle's Elixir decided to release the datasets as a way to allow members of the community to conduct anaylsis and start their esports career in some analyst or coaching positon.
</p>

<h4>If Zoe is banned, is her team more or less likely to win (or neither)?</h4>
<p>
    For starters, we'll ask a simple question. If the champion Zoe is banned, does this make her team more likely to win? Does it make a difference? First, we need to extract matches in which Zoe is banned. Let's keep blue and red bans separate for now, since this might make answering the question easier later.
    </p>
<b>
First, we'll get the matches when the blue team banned Zoe
    </b>

```python
blueban =  league[(league.side == "Blue") & (league.position == "Team")]
bluezoeban = blueban[((league.ban1 == "Zoe") | (league.ban2 == "Zoe") | (league.ban3 == "Zoe") | (league.ban4 == "Zoe") | (league.ban5 == "Zoe"))]
bluezoeban
```

```python
print ("Zoe was banned " + str(len(bluezoeban)) +  " times out of " + str(len(blueban)) + " Blue side games (" + str(len(bluezoeban)*100/len(blueban)) + "%)")
```

<b>And we'll do the same for the red team</b>

```python
redban =  league[(league.side == "Red") & (league.position == "Team")]
redzoeban = redban[((league.ban1 == "Zoe") | (league.ban2 == "Zoe") | (league.ban3 == "Zoe") | (league.ban4 == "Zoe") | (league.ban5 == "Zoe"))]
redzoeban
```

```python
print ("Zoe was banned " + str(len(redzoeban)) +  " times out of " + str(len(redban)) + " Red side games (" + str(len(redzoeban)*100/len(redban)) + "%)")
```

<h4>Now we want all the matches where Zoe was banned by at least one team, then we could get it with:</h4>

```python
print ("Zoe was banned " + str(len(redzoeban) + len(bluezoeban)) +  " times out of " + str(len(redban)) + " games (" + str((len(redzoeban) + len(bluezoeban))*100/len(redban)) + "%)")
```

<h4>Now let's look at the overall ban rate compared to total games played. </h4>

```python
total_games = league[league.position == 'Team']
num_games = len(total_games)/2
ban_rate =  (len(redzoeban) + len(bluezoeban))/num_games
print (ban_rate * 100)
```

<b>And if we wanted all the matches where Zoe was banned by both teams, we can do an intersection of the two </b>

```python
(redban & blueban).dropna()
```

```python
## The result should be to be expected because both teams cannot ever both
## ban the same champions especially in competitive. This is a nice check
## though.

## For red team bans, how many of those matches did the red team win?
redzoebanwin = redzoeban[redzoeban.result == 1]
print("When Zoe was banned on the Red side, they won " + str(len(redzoebanwin)) + " out of " + str(len(redzoeban)) + " games (" + str(len(redzoebanwin)*100/len(redzoeban)) + "%)") 
print("and lost " + str(len(redzoeban)-len(redzoebanwin)) + " out of " + str(len(redzoeban)) + " games (" + str(100-(len(redzoebanwin)*100/len(redzoeban))) + "%)")

## This would give us a dataframe of all the matches that the blue team 
## won among those where the red team banned Zoe
bluezoebanwin = bluezoeban[bluezoeban.result == 1]
print("When Zoe was banned on the Blue side, they won " + str(len(bluezoebanwin)) + " out of " + str(len(bluezoeban)) + " games (" + str(len(bluezoebanwin)*100/len(bluezoeban)) + "%)") 
print("and lost " + str(len(bluezoeban)-len(bluezoebanwin)) + " out of " + str(len(bluezoeban)) + " games (" + str(100-(len(bluezoebanwin)*100/len(bluezoeban))) + "%)")

# After the ban phase, the winrate is based more heavily on the team's execution of the game instead of solely on this one champion being banned
# because in all of these cases, the Zoe ban is the control variable because it is not changing in any of these cases
```

```python
## Who is the most banned character? Does banning this character make 
## your team more likely to win?

## We want to take the task some conventional wisdom that may be floating
## around the community about which characters you should ban if you want 
## to increase your chances to win. Let's see if this holds water.

## First we need to find the most banned character. Probably lots of ways
## to get at this, but one thign we can do is a groupy by each ban slot as
## before, for each building a dictionary giving banned champions as keys
## and the number of matches they were banned in as values:

allbans = league[(league.position == "Team")][league.columns[9:15]]
bans1 = allbans['ban1'].value_counts()
bans2 = allbans['ban2'].value_counts()
bans3 = allbans['ban3'].value_counts()
bans4 = allbans['ban4'].value_counts()
bans5 = allbans['ban5'].value_counts()
bans = bans1.add(bans2, fill_value=0).add(bans3, fill_value=0).add(bans4, fill_value=0).add(bans5, fill_value=0)
bans.sort_values(ascending=False)
bansdf = pd.DataFrame(bans).rename(columns={0: 'bancount'})
bansdf = bansdf.sort_values(by='bancount',ascending=False)
bansdf.head()
```

```python
## So we have our answer. The most banned champion across the major regions
## is Aatrox.

## Out of curiosity, what's the distribution of bans look like acrosss
## champions?
a4_dims = (12, 22)
fig, ax = plt.subplots(figsize=a4_dims)
sns.barplot(y=bansdf.index, x='bancount', data=bansdf, ax=ax)
```

```python
## Does banning Aatrox make you more likely to win?

## We'll do the same anaylsis we did for Zoe, but now for Aatrox,
## And we'll do the same for the red team
redban =  league[(league.side == "Red") & (league.position == "Team")]
redaatroxban = redban[((league.ban1 == "Aatrox") | (league.ban2 == "Aatrox") | (league.ban3 == "Aatrox") | (league.ban4 == "Aatrox") | (league.ban5 == "Aatrox"))]
redaatroxban
```

```python
blueban =  league[(league.side == "Blue") & (league.position == "Team")]
blueaatroxban = blueban[((league.ban1 == "Aatrox") | (league.ban2 == "Aatrox") | (league.ban3 == "Aatrox") | (league.ban4 == "Aatrox") | (league.ban5 == "Aatrox"))]
blueaatroxban
```

```python
## For red team bans, how many of those matches did the red team lose?

## Did red team win more often than they lost? Here are the matches red 
## team won
print ("Aatrox was banned " + str(len(redaatroxban)) +  " times out of " + str(len(redban)) + " Red side games (" + str(len(redaatroxban)*100/len(redban)) + "%)")
```

```python
## What do we get for the case where blue team bans Aatrox?

## Did blue team win more often than they lost? Here are the match blue won?
print ("Aatrox was banned " + str(len(blueaatroxban)) +  " times out of " + str(len(blueban)) + " Blue side games (" + str(len(blueaatroxban)*100/len(blueban)) + "%)")
```

```python
## Hmmm...maybe not a signal here...might not be much point overall to 
## banning Aatrox. However, it may be the case that if you ban Aatrox and perhaps 
## another character this improves your chances? That's an interesting 
## question for another time. Could it also be the case that banning 
## Aatrox levels the playing field?
```

```python
## In a game where Aatrox wasn't banned, was he picked? Was the team more 
## likely win?

## First we need matches where Aatrox was not banned. We'll get the games where Aatrox was picked.
## From there we will look at the dataframe and look at the games where it was picked and they won.
picked = league[league.champion == 'Aatrox']
print(str(len(picked)))
picked_win = picked[picked.result == 1]
str(len(picked_win) * 100/len(picked)) + "%"
```

```python
## As we can see from the data the most banned champion has a ~66% win rate when picked, which is substantially high considering
## that out of 91 games, there is still a 66% win rate, which gives indication that either the champion at the time of the season
## gives the team such an advantage that it swings their chances of winning more often than not. 
```

```python
## Now let's look at the overall ban rate compared to total games played for the most banned champion
picked = league[league.champion == 'Aatrox']
total_games = league[league.position == 'Team']
num_games = len(total_games)/2
picked_rate = (len(picked)/num_games)
print (picked_rate * 100)
ban_rate =  (len(blueaatroxban) + len(redaatroxban))/num_games
print (ban_rate * 100)
```

```python
## As we can see the most banned champion is banned  ~66% of the games and of those games that aren't there is a win rate of ~66%.
## If we take this into perspective Aatrox is picked approximately half of the time it's up if not banned and of those times,
## the team with Aatrox has a win rate of 66%. In addition, this makes Aatrox have 83% of games picked or banned.
```

```python
## Now let's look at the second most banned champion (Taliyah).
## And we'll do the same analysis
picked = league[league.champion == 'Taliyah']
print(str(len(picked)))
picked_win = picked[picked.result == 1]
str(len(picked_win) * 100/len(picked)) + "%"
```

```python
## As we can see from the data the second most banned champion has a 60% win rate when picked. If from these two data point we
## we any correlation between the most banned champions. We can see that often times those most banned have more than at least a
## 50% win rate causing teams that play against these champions to be a greater disadvantage of winning. 
```
